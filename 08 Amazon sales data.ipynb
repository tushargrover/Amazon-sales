```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
df_ama = pd.read_csv(r"C:\Users\Acer\Downloads\Amazon Sales Records (2).csv")
```

```python
df_ama
```

```python
df_ama.info()
```

```python
df_ama['Order Date'] = pd.to_datetime(df_ama['Order Date'])
df_ama['Order Date'] = df_ama['Order Date'].dt.strftime('%d/%m/%Y')
```

```python
df_ama['Ship Date'] = pd.to_datetime(df_ama['Ship Date'])
df_ama['Ship Date'] = df_ama['Ship Date'].dt.strftime('%d/%m/%Y')
```

```python
df_ama
```

```python
df_ama['Day'] = pd.DatetimeIndex(df_ama['Order Date']).day_name()

##df_airport['day_of_week'] = df_airport['Date'].dt.day_name() --- if time & date both are given in the same column
```

```python
df_ama
```

```python
df_year = df_ama['Order Date'].apply(lambda x: x[-4:])
```

```python
df_ama = pd.concat([df_ama, df_year], axis=1)
```

```python
df_ama
```

```python
df_ama.columns
```

```python
df_ama.columns = ['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority',
       'Order Date', 'Order ID', 'Ship Date', 'Units Sold', 'Unit Price',
       'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit', 'Day',
       'Year']
```

```python
df_ama[['Region','Units Sold']].groupby(['Region'])['Units Sold'].sum().sort_values().plot.bar(x= 'Region' , y = 'Units Sold')
plt.rcParams['figure.figsize'] = (15,6)
plt.title('Unit sold regionwise',fontsize=25)
plt.ylabel('Average of units sold', fontsize=15) 
plt.xlabel('Region', fontsize=25)
plt.show()

## sum of the sales
```

```python
sns.barplot(x='Region' , y='Units Sold', data=df_ama)

##mean
```

```python
df_ama[['Region','Units Sold']].groupby(['Region'])['Units Sold'].mean().sort_values().plot.bar(x= 'Region' , y = 'Units Sold')
plt.rcParams['figure.figsize'] = (15,6)
plt.title('Unit sold regionwise',fontsize=25)
plt.ylabel('Average of units sold', fontsize=15) 
plt.xlabel('Region', fontsize=25)
plt.show()

##mean
```

```python
df_ama[['Country','Total Revenue']].groupby(['Country'])['Total Revenue'].sum().plot.bar(x= 'Country' , y = 'Total Revenue')
plt.rcParams['figure.figsize'] = (15,6)
plt.title('Unit sold regionwise',fontsize=25)
plt.ylabel('Average of units sold', fontsize=15) 
plt.xlabel('Country', fontsize=25)
plt.show()
```

```python
df_ama[['Country','Total Revenue']].groupby(['Country'])['Total Revenue'].sum().plot.bar(x= 'Country' , y = 'Total Revenue')
plt.rcParams['figure.figsize'] = (15,6)
plt.title('Unit sold regionwise',fontsize=25)
plt.ylabel('Average of units sold', fontsize=15) 
plt.xlabel('Region', fontsize=25)
plt.show()
```

```python
df_ama[['Item Type', 'Total Profit']].groupby(['Item Type'])['Total Profit'].sum().plot.bar(x='Item Type', y= 'Total Profit')
plt.rcParams['figure.figsize'] = (15,8)
plt.title('Item wise profit', fontsize = 20)
plt.ylabel('Profit', fontsize =12)
plt.xlabel('Items', fontsize=12)
plt.show()
```

```python
df_ama[['Sales Channel', 'Total Revenue']].groupby(['Sales Channel'])['Total Revenue'].sum().plot.bar(x = 'Sales Channel', y='Total Revenue')
plt.rcParams['figure.figsize'] = (15,8)
plt.title('Sales channel wise revenue', fontsize = 20)
plt.ylabel('Revenue', fontsize =12)
plt.xlabel('Channel', fontsize=12)
plt.show()
```

```python
df_ama[['Day', 'Units Sold']].groupby(['Day'])['Units Sold'].sum().plot.bar(x = 'Day', y='Units Sold')
plt.rcParams['figure.figsize'] = (15,8)
plt.title('Daywise sales', fontsize = 15)
plt.ylabel('Sales', fontsize=12)
plt.xlabel('daywise', fontsize = 12)
plt.show()
```
